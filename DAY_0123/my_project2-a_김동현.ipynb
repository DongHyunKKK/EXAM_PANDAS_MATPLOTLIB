```python
import pandas as pd
import numpy as np
```

```python
file = '../DATA/2023빅콘테스트_어드밴스드리그_예술의전당.csv'
```

```python
artDF = pd.read_csv(file)
```

```python
artDF.head()
```

```python
# 장소는 콘서트홀, 할인형태가 공연진행석인 것은 제외
conDF = artDF[artDF['place'] == '콘서트홀']
conDF = conDF[conDF['discount_type'] != '공연진행석']
conDF.head()
```

```python
conDF['genre'].isna().sum()
```

```python
# df1은 장르 결측된 데이터
df1 = conDF[conDF['genre'].isna()]
```

```python
# df2는 장르 결측 안된 데이터
df2 = conDF.dropna(subset = ['genre'])
```

```python
# df1와 df2의 공연날짜 중 같은 날을 찾는다. 이 날에 했던 공연 결측 데이터를 채울 수 있다. 
set(df1['play_date']) & set(df2['play_date'])
```

```python
# df1와 df2의 같은 공연날짜에는 공연이 한 번 있었다.
for date in set(df1['play_date']) & set(df2['play_date']):
    print(set(df2[df2['play_date'] == date]['play_st_time']), set(df2[df2['play_date'] == date]['genre']))
```

```python
for i in conDF.index:
    if conDF['play_date'][i] == 20190714:
        conDF.loc[i, 'genre'] = '교향곡'      
    elif conDF['play_date'][i] == 20210101:
        conDF.loc[i, 'genre'] = '클래식' 
    elif conDF['play_date'][i] == 20221120:
        conDF.loc[i, 'genre'] = '클래식'
```

```python
conDF = conDF.dropna(subset = ['genre'])
```

```python
conDF['genre'].isna().sum()
```

```python
# 원본 데이터에서 장소가 콘서트홀이고 할인형태가 공연진행석인 것 그리고 장르가 결측된 샘플 중 일부 제외하고 나머지를 추출하여 csv파일로 변환
conDF.to_csv('conDF.csv', index = False)
```
