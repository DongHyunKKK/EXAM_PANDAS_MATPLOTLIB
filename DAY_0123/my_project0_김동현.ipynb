```python
bylawDF = pd.DataFrame()
for file in file_list:  # glob(file + '*.txt')는 리스트
    f = open(file, 'r')
    line_list = f.readlines()  # readlines() 함수는 줄 바꿈을 기준으로 각줄을 요소로 하는 리스트를 반환한다.
    f.close()
    
    df = pd.DataFrame(line_list).T
    df['context'] = df[list(range(2, df.shape[1]))].apply(' '.join, axis = 1)
    df = df[[1, 'context']]  # 첫 번째 열은 \n이므로 제외
    bylawDF = pd.concat([bylawDF, df], ignore_index = True)  # ignore_index = True로 설정하지 않으면 행 번호가 모두 0으로 나온다.

bylawDF.columns = ['name', 'context']  # 첫 번째 열 이름 바꾸기.
bylawDF.index = bylawDF.index + 1
bylawDF['name'] = bylawDF['name'].str.rstrip('\n')  # name열에 있는 각 데이터의 맨 끝에 있는 \n 제거
```
